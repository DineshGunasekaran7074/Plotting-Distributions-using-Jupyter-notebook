```python
sea.boxplot(carss['Engine_Cylinders'])
mat.title('Engine_Cylinders')
mat.show()
```

#### The boxplot shows that high number of cars have 4 to 6 engine cylinders and the ouliers shows that some number of cars have 1 cylinder, 10 cylinders, 12 cylinders and extreme outliers that a car has 16 engine cylinders 

```python
sea.boxplot(y=carss['Engine_Cylinders'])
mat.title('Engine_Cylinders')
mat.show()
```

#### Boxplot has been changed vertically to have a better understanding of the Engine cylinders of the cars in the dataset

```python
sea.jointplot('Engine_HP','city_mpg',carss,kind='reg',color='g')
mat.show()
```

#### Jointplot is created for two variables "Engine HP" and "city mpg"  the cars which has 80 to 650 horse power has  city mpg around 5 to 25 , the car with 1001 horse power has the city mpg of less than 1 city mpg, the car with 50 Hp has City mpg of 137.    

```python
carss=carss[(carss.Engine_Cylinders<300)&(carss.Popularity<2000)]
sea.jointplot('Engine_Cylinders','Popularity',carss,kind="resid",color='b')
mat.show()
```

#### Here cars with engine cylinders more than 300 are taken in innerplot and cars with popularity more than 2000 are taken on outerplot. The jointplot shows that the cars with cylinders 4 to 8 have popularity arounf 200 to 400

```python
carss=carss[(carss.Engine_Cylinders<300)&(carss.Popularity<1000)]
sea.jointplot(x='Engine_Cylinders',y='Popularity',data=carss,kind='hex',color='g')
mat.show()
```

#### Here jointplot is created for the variables 'Engine_Cylinders' and 'Popularity' with the value representaion by using hexagon on green color. The darkness of hexagon depends upon the the number of Engine Cylinders in every car and higher the darkness shows that high number of cylinders in that car. The darkest hexagon on the dataset shows that the car with 4 to 5 cylinders have the popularity of 900 (app).
